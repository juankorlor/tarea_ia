## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números: 
1. Grafique los números aleatorios,

```
#Para todas las simulaciones:
N=100000
m=3000
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```
p=0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
print("mean:", mean, "variance:",var, "skewness:",skew,"kurtosis", kurt)
r = bernoulli.rvs(p, size=N)
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print(r.argmax())
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()

plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")

avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```
# binomial con cierto p entre 0 y 1 y cierto n
nb,pb = 10, 0.5
Nb=N
mb=100

from scipy.stats import binom
meanb, varb, skewb, kurtb = binom.stats(nb, pb, moments='mvsk')
print("media:", meanb, "varianza:", varb, "asimetría:", skewb, "curtosis:", kurtb)

rb= binom.rvs(nb, pb, size=Nb)
plt.plot(rb)
v_eb=rb.var()
m_eb=rb.mean()

plt.title(str(Nb)+ " Puntos distribuidos Binomial con p= " + str(pb) + " y n= " + str(nb))
plt.show()

rb.sort()
plt.plot(rb)
plt.title(str(Nb)+" Puntos distribuidos Binomial ordenados con p="+str(pb)+ " y n= "+ str(nb))
plt.show()

plt.hist(rb, bins=mb)
plt.title("Histograma de los "+str(Nb)+" puntos distribuidos binomial con p="+str(pb) + " y n= "+ str(nb))
plt.show()

print("Varianza teórica=", varb, "media teórica=", meanb)
print("Varianza estimada=", v_eb, "media estimada=", m_eb)
```

```
# Geométrica con cierto p entre 0 y 1
pG = 0.3
NG=N
mG=mb

from scipy.stats import geom
meanG, varG, skewG, kurtG = geom.stats(pG, moments='mvsk')
print("media:", meanG, "varianza:", varG, "asimetría:", skewG, "curtosis:", kurtG)
rG= geom.rvs(pG, size=NG)
plt.plot(rG)
m_eG=rG.mean()
v_eG=rG.var()

plt.title(str(NG)+ " puntos distribuidos Geometrica con p= " + str(pG))
plt.show()

rG.sort()
plt.plot(rG)
plt.title(str(NG)+" puntos distribuidos Geometrica ordenados con p="+str(pG))
plt.show()

plt.hist(rG, bins=mG)
plt.title("Histograma de los "+str(NG)+" puntos distribuidos Geometrica con p="+str(pG))
plt.show()

print("Varianza teórica=", varG, "media teórica=", meanG)
print("Varianza estimada=", v_eG, "media estimada=", m_eG)
```

```
# Poisson con cierto Lamnda L
L = 5
Np=N
mp=100

from scipy.stats import poisson
meanp, varp, skewp, kurtp = poisson.stats(L, moments='mvsk')
print("media:", meanp, "varianza:", varp, "asimetría:", skewp, "curtosis:", kurtp)
rp= poisson.rvs(L, size=Np)
plt.plot(rp)
m_ep=rp.mean()
v_ep=rp.var()

plt.title(str(Np)+ " puntos distribuidos Poisson con L= " + str(L))
plt.show()

rp.sort()
plt.plot(rp)
plt.title(str(Np)+" puntos distribuidos Poisson ordenados con L="+str(L))
plt.show()

plt.hist(rp, bins=mp)
plt.title("Histograma de los "+str(Np)+" puntos distribuidos Poisson con L="+str(L))
plt.show()

print("Varianza teórica=", varp, "media teórica=", meanp)
print("Varianza estimada=", v_ep, "media estimada=", m_ep)
```

```
# Uniforme entre 0 y 255
NC=1000
mC=300
from scipy.stats import uniform 
meanU, varU, skewU, kurtU = uniform.stats(moments='mvsk')
rU=uniform.rvs(loc=0, scale=255, size=NC)
plt.plot(rU)
m_eU=rU.mean()
v_eU=rU.var()

plt.title(str(NC)+" puntos distribuidos uniformemente entre 0 y 255")
plt.show()

rU.sort()
plt.plot(rU)
plt.title(str(NC)+" puntos distribuidos uniformes ordenados")
plt.show()


plt.hist(rU, bins=mC)
plt.title("Histograma de los "+str(NC)+" puntos distribuidos uniformemente  entre 0 y 255")
plt.show()

print("Varianza teórica=", varU, "media teórica=", meanU)
print("Varianza estimada=", v_eU, "media estimada=", m_eU)
```
